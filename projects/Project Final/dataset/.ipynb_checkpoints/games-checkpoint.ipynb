```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data_vgames.csv')
```

```python
df.head(3)
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df.describe()
```

```python
df.loc[df['Year']==2000].groupby('Genre')['Global_Sales'].sum().sort_values().plot.bar()
```

```python
df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10).plot(kind='barh')
```

```python
df.groupby('Name')['Global_Sales'].sum().sort_values(ascending=False).head(20).plot(kind='barh')
```

```python
df.groupby('Year')['NA_Sales','EU_Sales', 'JP_Sales', 'Other_Sales'].sum().plot(kind='barh', stacked=True, figsize=(16,10)).legend(bbox_to_anchor=(1,1))
```
